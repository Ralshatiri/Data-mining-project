```python
# Handle missing values
# -------------------------------

# Display the missing values in each column before removing missing values
missing_values = Diabetes_DataFrame.isnull().sum()
print("The missing values before handling:")
print(missing_values)

# replace missing values with the mean on BMI (4 missing value)
Diabetes_DataFrame["BMI"] =Diabetes_DataFrame["BMI"].fillna(Diabetes_DataFrame["BMI"].mean(numeric_only=True))

# Counting the number of female and males in dataset
count_female =0
for index,row in Diabetes_DataFrame.iterrows():
    if row["Gender"] =="Female":
     count_female=count_female+1

print(f"The number of females is {count_female}")

count_male =0
for index,row in Diabetes_DataFrame.iterrows():
    if row["Gender"] =="Male":
     count_male=count_male+1

print(f"The number of male is {count_male}")

# Removing the rows that has a missing values in Pregancies (42 missing value)
Diabetes_DataFrame = Diabetes_DataFrame.dropna(subset=["Pregancies"])

# remove one row that has a missing class label (1 missing value)
Diabetes_DataFrame = Diabetes_DataFrame.dropna(subset=["Diabetic"])


# replace no with 0 in Pdiabetes column
Diabetes_DataFrame["Pdiabetes"]=Diabetes_DataFrame["Pdiabetes"].replace("no",0)
# replace yes with 1 in Pdiabetes column
Diabetes_DataFrame["Pdiabetes"]=Diabetes_DataFrame["Pdiabetes"].replace("yes",1)

# fill the missing values in Pdiabetes column with mode (1 missing value)
Diabetes_DataFrame["Pdiabetes"] =Diabetes_DataFrame["Pdiabetes"].fillna(Diabetes_DataFrame["Pdiabetes"].mode()[0])

# Display the missing values in each column after removing/ replacing missing values
missing_values = Diabetes_DataFrame.isnull().sum()
print("The missing values after handling:")
print(missing_values)
```

**Normalization**


---


- The code first identifies the columns to be normalized, extracts the data from those columns, creates a Min-Max scaler, fits the scaler to the data, and transforms the data to a new range (usually between 0 and 1). Finally, the normalized values replace the original values in the specified columns.



```python
# Normalization
#  -------------------------------



from sklearn.preprocessing import MinMaxScaler

# select columns to normalize
columns_to_normalize = ["BMI","Sleep","SoundSleep"]
data_to_normalize =Diabetes_DataFrame[columns_to_normalize]

# normalize the numerical values using min-max
minmax_scaler= MinMaxScaler()
normalized_data_minmax = minmax_scaler.fit_transform(data_to_normalize)
Diabetes_DataFrame[columns_to_normalize] =normalized_data_minmax


# display columns after normalization

# BMI
print("BMI:")
print(Diabetes_DataFrame["BMI"],"\n")

# Sleep
print("Sleep:")
print(Diabetes_DataFrame["Sleep"],"\n")

# SoundSleep
print("SoundSleep:")
print(Diabetes_DataFrame["SoundSleep"],"\n")

```
